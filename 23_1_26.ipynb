Q5. Create a dictionary where:
• Keys = student names
• Values = marks (integer)
Write a menu-based program where user presses a key (’A’, ‘B’, ‘C’, ‘D’)
depending on the operation they want to perform on the dictionary:
1. A - Add a student
2. B - Update marks
3. C - Search for a student
4. D - Display all students and marks


```python
from ast import Break
dic={
    "muskan":100,
    "vikas":55,
    "viso": 75,
    "guddo":85,
    "rammu":44
}
while(True):
  menu=input("enter menu: ")
  if(menu=="A"):
    name=input("enter name:")
    marks=int(input("enter marks:"))
    dic[name]=marks
    print("succesfully added")

  elif(menu=="B"):
    name=input("enter update name:")
    if name in dic:
      marks=int(input("new marks: "))
      dic[name]=marks
      print("succesfully added")
    else:
      print("no student found")

  elif(menu=="C"):
    name=input("enter name to search: ")
    if name in dic:
      print(f"marks of {name} is {dic[name]}")
    else:
      print("invalis name")
  elif(menu=="D"):
    print(dic)

  elif(menu=="quit"):
      break
  else:
    print("invalid menu")
```

Create a dictionary that maps each word to its length.
Example:
{"apple": 5, "banana": 6, "kiwi": 4, ...}

Write a program that takes a string from the user and prints the number of
spaces in the string.

```python
s = input("Enter a string: ")
count = 0

for ch in s:
    if ch == " ":
        count += 1

print("Number of spaces:", count)
```

Write a program to check whether two lists share no common elements.
# share no common elements list1 = [1, 2, 3, 4] list2 = [5, 6, 7, 8]
# share common elements list1 = [1, 2, 3] list2 = [3, 4]


```python
list1=[1,2,3,4,5]
list2=[3,4,5,6]
if set(list1).intersection(set(list2)):
  print("share common elements")
  print(set(list1).intersection(set(list2)))
else:
  print("no common elements")
```

9. Given a list, print all elements that appear more than once in the list.
[Hint - use sets]

```python
list1=[1,2,3,4,5,7,2,5,6,7]
seen=set()
duplicate=set()
for num in list1:
  if(num in seen):
    duplicate.add(num)
  else:
    seen.add(num)
print(f"all elements that appear more than once in the list: {duplicate}")
```
