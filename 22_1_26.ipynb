```python
#factorial
def fact(n):
  ans=1
  for i in range(1,n+1):
    ans*=i
  return ans

fact(10)
```

Write a function that takes two integers and and prints all even
numbers between them (inclusive).

```python
def even_btw(a,b):
  for i in range(a,b+1):
    if(i%2==0):
      print(i)
even_btw(2,22)
```

Write a function to return the count the number of digits in a number, n

```python
def countNum(n):
  count=0
  for i in n:
    count+=1
  return count

print(countNum([1, 2, 3, 4]))     # 4
print(countNum("Muskan"))        # 6
print(countNum((10, 20, 30)))    # 3
```

Write a program to print all numbers from 1 to 100 that are divisible by both 3
and 5.

```python
for i in range(1,100+1):
  if(i%3==0 and i%5==0):
    print(i)
```

Write a function that returns if is a prime number and
otherwise, using a loop.


```python
def isprime(n):
  for i in range(2,n):
    if(n%i==0):
      return False
  return True

isprime(83)
```
